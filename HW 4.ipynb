## Functions for flipping and non linearities

```python
def flip_left_right(image):
    W, H, C = image.shape
    flipped_image = np.zeros((W, H, C))
    for i in range(H):
        flipped_image[:, H-i-1, :] = image[:, i, :]

    return flipped_image

def flip_up_down(image):
    W, H, C = image.shape
    flipped_image = np.zeros((W, H, C))
    for i in range(H):
        flipped_image[H-i-1, :, :] = image[i, :, :]

    return flipped_image

# defining the non-linearities
def sigmoid(X):
    return 1/(1 + np.exp(-X))

def tanh(X):
    return (np.exp(X) - np.exp(-X))/(np.exp(X) + np.exp(-X))

def ReLU(X):
    return (X + np.abs(X))/2

def PReLU(X):
    a = 0.2    # this is a hyperparameter, set as 0.2 for this homework
    m, n = X.shape
    for i in range(m):
        for j in range(n):
            if X[i, j] < 0:
                X[i, j] *= a
    return X
```

## Convolution and Pooling functions

```python
'''
Takes in an image with dimensions W x H x C and convolves it with an input kernel K x K x C, gives J = I*h

# Inputs:
img: A numpy array of size W x H x C
ker: A numpy array of size K x K x C
stride: int (default = 1)
padding: bool (default = True)
non_lin: string (default = 'sigmoid'; available choices: ['sigmoid', 'tanh', 'ReLU', 'PReLU'])

# Outputs:
kernel and the activation map after convolution

# Returns:
Output activation map 
'''


def convolve_2D (img, ker, stride=1, padding=False, non_lin='ReLU'):
    
    n_lins = ['sigmoid', 'tanh', 'ReLU', 'PReLU']
    if non_lin not in n_lins:
        raise ValueError("Available non-linearities are:", n_lins)
        
    stri = stride-1
    W, H, C = img.shape
    K, K, C = ker.shape

    # displaying the kernel
    # flipping the kernel for convolution
    ker = flip_up_down(flip_left_right(ker))
    
    # setting the dimensions for the padded and non-padded outputs
    W_out_pad = int((W + 2*(K-1) - K)/stride + 1)
    H_out_pad = int((H + 2*(K-1) - K)/stride + 1)
    W_out = int((W-K)/stride + 1)
    H_out = int((H-K)/stride + 1)
    
    
    if padding:
        convolvedImage = np.zeros((W_out_pad, H_out_pad))
    else:
        convolvedImage = np.zeros((W_out, H_out))

    if padding:
        paddedImage = np.zeros((W + 2*(K-1), H + 2*(K-1), C))
        for i in range(W):
            for j in range(H):
                paddedImage[(K-1)+i, (K-1)+j] = img[i, j]
    
        temp_ud = flip_up_down(img)
        for i in range(0, K-1):
            for j in range(K-1, H+K-1): 
                paddedImage[K-2-i, j] = temp_ud[W-i-1, j-(K-1)]         # top
                paddedImage[W+K-1+i, j] = temp_ud[(K-1)-i, j-(K-1)]     # bottom

        temp_lr = flip_left_right(img)
        for i in range(K-1, W+K-1):
            for j in range(0, K-1):
                paddedImage[i, K-2-j] = temp_lr[i-(K-1), H-j-1]         #left
                paddedImage[i, H+K-1+j] = temp_lr[i-(K-1), j]           #right
    

        # corners
        # numbering the corners 1, 2, 3, 4 in the clockwise manner starting from the top-left corner and picking up image parts for each corner
        corner1 = np.zeros((K-1, K-1, C))
        corner2 = np.zeros((K-1, K-1, C))
        corner3 = np.zeros((K-1, K-1, C))
        corner4 = np.zeros((K-1, K-1, C))
    
        # getting the corners from the image
        for i in range(0, K-1):
            for j in range(0, K-1):
                corner1[i, j] = img[i, j]
                corner2[i, j] = img[i, (H-(K-1))+j]
                corner3[i, j] = img[(W-(K-1))+i, (H-(K-1))+j]
                corner4[i, j] = img[(W-(K-1))+i, j]
    
        # mirroring the corners
        for i in range(C):
            corner2[:, :, i] = corner2[:, :, i].T
            corner3[:, :, i] = corner3[:, :, i].T

        J = np.zeros((K-1, K-1), dtype=int)
        for i in range(K-1):
            for j in range(K-1):
                if i + j + 1 == (K-1):
                    J[i, j] = 1
        for i in range(C):
            corner1[:, :, i] = ((corner1[:, :, i]@J).T)@J
            corner4[:, :, i] = ((corner4[:, :, i]@J).T)@J

        # replacing the mirrored matrices in the corners
        for i in range(0, K-1):
            for j in range(0, K-1):
                paddedImage[i, j] = corner1[i, j]
                paddedImage[i, H+K-1+j] = corner2[i, j]
                paddedImage[W+K-1+i, H+K-1+j] = corner3[i, j]
                paddedImage[W+K-1+i, j] = corner4[i, j]
        
        
    # applying convolution
        for i in range(W_out_pad):
            for j in range(H_out_pad):
                temp_img = paddedImage[i+(stri*i):(i+(stri*i)+K), j+(stri*j):(j+(stri*j)+K), :]
                for k in range(C):
                    convolvedImage[i, j] += (np.sum(ker[:, :, k]*temp_img[:, :, k]))
                

    
    else:
        paddedImage = np.array(img)
        for i in range(W_out):
            for j in range(H_out):
                temp_img = paddedImage[i+(stri*i):(i+(stri*i)+K), j+(stri*j):(j+(stri*j)+K), :]
                for k in range(C):
                    convolvedImage[i, j] += (np.sum(ker[:, :, k]*temp_img[:, :, k]))
            
    # extracting the same dimensional image from the convolved image
    if padding and stride==1:
        final_image = convolvedImage[int((K-1)/2):(W + int((K-1)/2)), int((K-1)/2):(H + int((K-1)/2))]
    else:
        final_image = convolvedImage
    
    # normalising 
    final_image = final_image/(np.max(final_image) + 0.01)
    
    # applying the non-linearity to the convolution performed
    if non_lin == 'sigmoid':
        final_image = sigmoid(final_image)

    if non_lin == 'tanh':
        final_image = tanh(final_image)
    
    if non_lin == 'ReLU':
        final_image = ReLU(final_image)
    
    if non_lin == 'PReLU':
        final_image = PReLU(final_image)
    

    return final_image
```

```python
'''
Takes an input activation map with dimensions W x H and returns pooled output specific to the stride set as input

# inputs:
img: Numpy array of size W x H
stride: integer (default=1)
filt_size: filter size, integer (default=2)
pool_func: Function used in pooling (default='max'; available:['max', 'min', 'avg', 'norm']

# outputs:
input activation map,
activation map after pooling

# returns:
activation map after pooling

'''

def pooling(img, stride=1, filt_size=2, pool_func='max'):
    
    
    # checking if pooling function is valid
    pool_funcs = ['max', 'min', 'avg', 'norm']
    if pool_func not in pool_funcs:
        raise ValueError('Input pool_func not valid. Valid choices are:', pool_funcs)
    
    # normalising the input activation map
    img = img/(np.max(img) + 0.01)
    

    # setting stri as stride-1 for computational ease
    stri = stride-1
    
    # setting dimensions for the pooled output
    W, H = img.shape
    W_out = int(np.ceil((W - filt_size + 1)/stride))
    H_out = int(np.ceil((H - filt_size + 1)/stride))

    pooled_out = np.zeros((W_out, H_out))

    # pooling acording to the specified pooling function
    if pool_func=='max':
        for i in range(W_out):
            for j in range(H_out):
                pooled_out[i, j] = np.max(img[i+(stri*i):i+(stri*i)+filt_size, j+(stri*j):j+(stri*j)+filt_size])
    
    if pool_func=='min':
        for i in range(W_out):
            for j in range(H_out):
                pooled_out[i, j] = np.min(img[i+(stri*i):i+(stri*i)+filt_size, j+(stri*j):j+(stri*j)+filt_size])
    
    if pool_func=='avg':
        for i in range(W_out):
            for j in range(H_out):
                pooled_out[i, j] = np.average(img[i+(stri*i):i+(stri*i)+filt_size, j+(stri*j):j+(stri*j)+filt_size])
    
    if pool_func=='norm':
        for i in range(W_out):
            for j in range(H_out):
                pooled_out[i, j] = np.linalg.norm(img[i+(stri*i):i+(stri*i)+filt_size, j+(stri*j):j+(stri*j)+filt_size])
    
    
    return pooled_out
```

## Layer functions implementing the convolution and pooling functions

```python
# uses the convolve_2D function to perform convolution on an image with 'm' kernels
# returns an activation volume

def convolution_layer(input, kernels, stride=1, padding=False, non_lin='ReLU'):
    
    # getting the number of kernels
    num_ker = kernels.shape[3]
    
    # checking if the number of channels in the input are same as the channels in the kernels
    if input.shape[2] != kernels.shape[2]:
        raise ValueError('Channels of input and kernels do not match.')


    fin_out = []
    for k in range(num_ker):
        fin_out.append(convolve_2D(input, kernels[:, :, :, k], stride, padding, non_lin))

    # creating the output volume
    w_out, h_out = fin_out[0].shape
    img_out = np.zeros((w_out, h_out, num_ker))
    for k in range(num_ker):
        img_out[:, :, k] = fin_out[k]

    return img_out
```

```python
# uses the pooling function to pool an input activation volume to the specified filter size, stride and pooling function
# returns the pooled volume of the input activation

def pooling_layer(input, filt_size=2, stride=1, pool_func='max'):
    
    out = []
    m = input.shape[2]
    for k in range(m):
        out.append(pooling(input[:, :, k], stride, filt_size, pool_func))

    w_out, h_out = out[0].shape
    img_out = np.zeros((w_out, h_out, m))
    for k in range(m):
        img_out[:, :, k] = out[k]

    return img_out
```

## Function for flattening the output of the convolution layers

```python
# function for flattening (unravelling) the input volume 
# returns a flattened output of size same as the given 'input' 

def flatten(input):
    w, h, m = input.shape
    out = np.zeros((m, w*h))
    for i in range(m):
        out[i] = input[:, :, i].reshape(w*h)
    out = out.reshape(w*h*m)
    
    weights = np.identity(out.shape[0])        # random weights
    return weights@out
```

## MLP function

```python
'''
MLP with 1 hidden layer node and 49 nodes in the hidden layer

Takes an input vector and gives an output vector for specified number of nodes. 

# Inputs:
input: 1D numpy array,
out_size: size of the output vector, integer
alpha: 1D numpy array, weights for the hidden layer
beta: 1D numpy array, weights for the output layer
non_lin: non-linearity to be applied, str=['sigmoid', 'tanh', 'ReLU', 'PReLU'], default='sigmoid'
softmax: to apply softmax or not, boolean, default: True

# returns
output vector with the specified size.
'''

def mlp (input, alpha, beta, non_lin='ReLU', softmax=True):

    # dimensions of alpha are: [input+1 x n_hlnodes]
    # dimensions of beta are: [n_hlnodes x output]
    input = input.reshape((input.shape[0], 1))
    
    # getting the biases
    bias_alph = alpha[0].T
    bias_alph = bias_alph.reshape((bias_alph.shape[0], 1))
    alpha = alpha[1:]

    # getting the hidden layer output
    out = alpha.T@input + bias_alph
    
    # applying non-linearities
    if non_lin == 'sigmoid':
        out = sigmoid(out)
    if non_lin == 'tanh':
        out = tanh(out)
    if non_lin == 'ReLU':
        out = ReLU(out)
    if non_lin == 'PReLU':
        out = PReLU(out)

    h_out = out

    # getting the biases out
    bias_beta = beta[0].T
    bias_beta = bias_beta.reshape((bias_beta.shape[0], 1))
    beta = beta[1:]

    # final output
    f_out = beta.T@out + bias_beta

    f_out = f_out/np.max(f_out)
    # applying softmax
    if (softmax == True):
        return h_out, np.exp(f_out)/(np.sum(np.exp(f_out)))
    else:
        return h_out, f_out
```

## Function to create random filter kernels for convolution

```python
# creates random lernels according to the input dimension, number and number of channels
def kernel_create(ker_size, n_kernels, n_channels):
    kernels = np.zeros((ker_size, ker_size, n_channels, n_kernels))
    for i in range(n_kernels):
        kernels[:, :, :, i] = np.random.uniform(-0.2, 0.2, size=(ker_size, ker_size, n_channels))

    return kernels
```

## Feed Forward function

```python
# performs a feedforward path of the CNN

def feed_forward(img, kernel_l1, kernel_l2, alpha, beta):
    
    # if the input image has only 2 dimensions i.e. no color channels, convert it to 3D numpy array showing only one channel i.e. B&W
    if len(img.shape) != 3:
        img = img.reshape((img.shape[0], img.shape[1], 1))

    img = np.array(img)
    
    in_conv_1 = img
    out = convolution_layer(img, kernel_l1, padding=False, non_lin='ReLU')
    in_pool_1 = out
    out = pooling_layer(out, 2, 2, 'max')
    in_conv_2 = out
    out = convolution_layer(out, kernel_l2, padding=False, non_lin='ReLU')
    in_pool_2 = out
    out = pooling_layer(out, 2, 2, 'max')
    out = flatten(out)
    in_mlp = out
    hid_out, fin_out = mlp(out, alpha, beta)
    return fin_out, hid_out, in_mlp, in_pool_2, in_conv_2, in_pool_1, in_conv_1
```

## Backprop for all the layers

```python
def mlp_backprop(input, alpha, beta, h_out, y_pred, y_act):

    # accounting the bias update
    temp_h_out = np.vstack((1, h_out))
    
    d_l_d_b = temp_h_out @ (y_act * (1-y_pred)).T
    d_l_d_b = d_l_d_b/(np.max(d_l_d_b)+0.01)

    d_l_d_z = beta[1:] @ (y_act * (1-y_pred))
    d_l_d_z = d_l_d_z/(np.max(d_l_d_z)+0.01)
    
    # computing the differential for the non-linearity
    d_nl_d_a = [1 if x>=0 else 0 for x in h_out]
    d_nl_d_a = np.array(d_nl_d_a)
    
    # accounting the bias update
    input = input.reshape((input.shape[0], 1))
    temp_input = np.vstack((1, input))

    d_l_d_a = temp_input @ (d_l_d_z * d_nl_d_a.reshape((d_nl_d_a.shape[0], 1)) * h_out).T
    d_l_d_a = d_l_d_a/(np.max(d_l_d_a)+0.01)

    d_l_d_x = alpha[1:] @ (d_l_d_z * d_nl_d_a.reshape((d_nl_d_a.shape[0], 1)) * h_out)
    d_l_d_x = d_l_d_x/(np.max(d_l_d_x) + 0.01)

    # computing the dims of the output gradient map
    dim = int(np.ceil(input.shape[0]**(1./3)))
    
    # reshaping the gradient map according to the input activation map
    d_l_d_x = d_l_d_x.reshape((dim, dim, dim))
    
    return d_l_d_a, d_l_d_b, d_l_d_x
```

```python
def pooling_backprop(input, prev_gradient, pool_func='max', stride=2, filt_size=2):
    stri = stride-1
    if len(input.shape) < 3:
        input = input.reshape((input.shape[0], input.shape[1], 1))

    W, H, C = input.shape[0], input.shape[1], input.shape[2]
    
    patch = np.zeros((filt_size, filt_size))
    gradient = np.zeros((W, H, C))
    for k in range(prev_gradient.shape[2]):
        for i in range(prev_gradient.shape[0]):
            for j in range(prev_gradient.shape[1]):
                patch += input[i+(stri*i):i+(stri*i)+filt_size, j+(stri*j):j+(stri*j)+filt_size, k]
                if patch[0][0]==np.max(patch):
                    gradient[i+(stri*i), j+(stri*j), k] = prev_gradient[i, j, k]
                
                if patch[0][1]==np.max(patch):
                    gradient[i+(stri*i), j+(stri*j)+1, k] = prev_gradient[i, j, k]
                
                if patch[1][0]==np.max(patch):
                    gradient[i+(stri*i)+1, j+(stri*j), k] = prev_gradient[i, j, k]

                if patch[1][1]==np.max(patch):
                    gradient[i+(stri*i)+1, j+(stri*j)+1, k] = prev_gradient[i, j, k]
    
    return gradient
    
```

```python
def conv_backprop(input, prev_gradient, kernels, stride=1):
    stri = stride-1
    # if there is only 1 filter kernel we make it into a 4D numpy array
    if len(kernels.shape) < 4:
        kernels = kernels.reshape((kernels.shape[0], kernels.shape[1], kernels.shape[2], 1))
        
    # getting the required shapes
    W_inp, H_inp, C_inp = input.shape
    W_ker, H_ker, C_ker = kernels.shape[0:3]

    if C_inp != C_ker:
        raise ValueError('The chanels for the kernels and the input activation map do not match.')
    
    # defining the updates for the kernels
    kernel_updates = np.zeros(kernels.shape)
    
    W_grad, H_grad, N_grad = prev_gradient.shape

    prev_gradient = ReLU(prev_gradient)
    
    
    for n in range(N_grad):
        for i in range(W_inp - W_ker + 1):
            for j in range(H_inp - H_ker + 1):
                patch = input[i+(stri*i):(i+(stri*i)+W_ker), j+(stri*j):(j+(stri*j)+H_ker), :]
                for k in range(C_inp):
                    kernel_updates[:, :, k, n] += patch[:, :, k] * prev_gradient[i, j, n]
                
    padded_gradient = np.zeros((W_grad + 2*(W_ker-1), H_grad + 2*(H_ker-1), N_grad))
    for i in range(W_grad):
        for j in range(H_grad):
            padded_gradient[(W_ker-1)+i, (H_ker-1)+j] = prev_gradient[i, j]
    
    temp_ud = flip_up_down(prev_gradient)
    for i in range(0, W_ker-1):
        for j in range(W_ker-1, H_grad+H_ker-1): 
            padded_gradient[W_ker-2-i, j] = temp_ud[W_grad-i-1, j-(H_ker-1)]         # top
            padded_gradient[W_grad+W_ker-1+i, j] = temp_ud[(W_ker-1)-i, j-(H_ker-1)]     # bottom

    temp_lr = flip_left_right(prev_gradient)
    for i in range(W_ker-1, W_grad+W_ker-1):
        for j in range(0, W_ker-1):
            padded_gradient[i, H_ker-2-j] = temp_lr[i-(W_ker-1), H_grad-j-1]         #left
            padded_gradient[i, H_grad+H_ker-1+j] = temp_lr[i-(W_ker-1), j]           #right


    # corners
    # numbering the corners 1, 2, 3, 4 in the clockwise manner starting from the top-left corner and picking up image parts for each corner
    corner1 = np.zeros((W_ker-1, H_ker-1, N_grad))
    corner2 = np.zeros((W_ker-1, H_ker-1, N_grad))
    corner3 = np.zeros((W_ker-1, H_ker-1, N_grad))
    corner4 = np.zeros((W_ker-1, H_ker-1, N_grad))

    # getting the corners from the image
    for i in range(0, W_ker-1):
        for j in range(0, H_ker-1):
            corner1[i, j] = prev_gradient[i, j]
            corner2[i, j] = prev_gradient[i, (H_grad-(W_ker-1))+j]
            corner3[i, j] = prev_gradient[(W_grad-(W_ker-1))+i, (H_grad-(H_ker-1))+j]
            corner4[i, j] = prev_gradient[(W_grad-(W_ker-1))+i, j]

    # mirroring the corners
    for i in range(C_ker):
        corner2[:, :, i] = corner2[:, :, i].T
        corner3[:, :, i] = corner3[:, :, i].T

    J = np.zeros((W_ker-1, H_ker-1), dtype=int)
    for i in range(W_ker-1):
        for j in range(H_ker-1):
            if i + j + 1 == (W_ker-1):
                J[i, j] = 1
    for i in range(C_ker):
        corner1[:, :, i] = ((corner1[:, :, i]@J).T)@J
        corner4[:, :, i] = ((corner4[:, :, i]@J).T)@J

    # replacing the mirrored matrices in the corners
    for i in range(0, W_ker-1):
        for j in range(0, H_ker-1):
            padded_gradient[i, j] = corner1[i, j]
            padded_gradient[i, H_grad+H_ker-1+j] = corner2[i, j]
            padded_gradient[W_grad+W_ker-1+i, H_grad+H_ker-1+j] = corner3[i, j]
            padded_gradient[W_grad+W_ker-1+i, j] = corner4[i, j]

    kernels_conv = kernels
    for n in range(kernels.shape[3]):
        kernels_conv[:, :, :, n] = flip_up_down(flip_left_right(kernels[:, :, :, n]))
    
    padded_gradient = padded_gradient/(np.max(padded_gradient) + 0.01)
    
    gradient = np.zeros(input.shape)
    temp_grad_conv = np.zeros((input.shape[0], input.shape[1]))
    for n in range(input.shape[2]):
        for i in range(input.shape[0]):
            for j in range(input.shape[1]):
                temp_grad = padded_gradient[i+(stri*i):(i+(stri*i)+W_ker), j+(stri*j):(j+(stri*j)+H_ker), :]
                for k in range(N_grad):
                    temp_grad_conv[i, j] += np.sum(kernels_conv[:, :, n, k] * temp_grad[:, :, k])

        gradient[:, :, n] = temp_grad_conv
    kernel_updates = kernel_updates/(np.max(kernel_updates) + 0.01)
    
    return kernel_updates, gradient
```

## SGD Variants

```python
def momentum(alpha, beta, kernel_l1, kernel_l2, alpha_update, beta_update, kernel_l1_update, kernel_l2_update, v_alph, v_beta, v_k1, v_k2, a, learning_rate):
    
    # calculating the velocities
    v_alph = a*v_alph - learning_rate*alpha_update
    v_beta = a*v_beta - learning_rate*beta_update
    v_k1 = a*v_k1 - learning_rate*kernel_l1_update
    v_k2 = a*v_k2 - learning_rate*kernel_l2_update

    # updating the weights
    alpha -= v_alph
    beta -= v_beta
    kernel_l1 -= v_k1
    kernel_l2 -= v_k2

    return alpha, beta, kernel_l1, kernel_l2, v_alph, v_beta, v_k1, v_k2
```

```python
def rmsprop(alpha, beta, kernel_l1, kernel_l2, alpha_update, beta_update, kernel_l1_update, kernel_l2_update, g_alph, g_beta, g_k1, g_k2, learning_rate):
    
    rho = 0.9
    delta = 0.00001
    
    g_alph = rho*g_alph + (1-rho)*(alpha_update*alpha_update) 
    alpha += (learning_rate/(np.sqrt(delta + g_alph)))*alpha_update
    
    g_beta = rho*g_beta + (1-rho)*(beta_update*beta_update) 
    beta += (learning_rate/(np.sqrt(delta + g_beta)))*beta_update
    
    g_k1 = rho*g_k1 + (1-rho)*(kernel_l1_update*kernel_l1_update) 
    kernel_l1 += (learning_rate/(np.sqrt(delta + g_k1)))*kernel_l1_update
    
    g_k2 = rho*g_k2 + (1-rho)*(kernel_l2_update*kernel_l2_update) 
    kernel_l2 += (learning_rate/(np.sqrt(delta + g_k2)))*kernel_l2_update

    return alpha, beta, kernel_l1, kernel_l2, g_alph, g_beta, g_k1, g_k2
```

## Training the CNN

```python
def train_test(input_train, input_test, batch_size, epochs, learning_rate, variant):

    if variant not in ['Vanilla', 'Momentum', 'RMSProp']:
        raise ValueError("Choose a valid variant from:", ['Vanilla', 'Momentum', 'RMSProp'])
    
    print('Training on the MNIST Dataset...')
    
    # initialising the weights
    kernel_l1 = kernel_create(5, 4, 1)
    kernel_l2 = kernel_create(5, 4, 4)
    beta = np.random.uniform(-0.2, 0.2, size=(50, 10))
    alpha = np.random.uniform(-0.2, 0.2, size=(65, 49))
    
    # updates for the weights
    update_alpha = np.zeros(alpha.shape)
    update_beta = np.zeros(beta.shape)
    update_kernel1 = np.zeros(kernel_l1.shape)
    update_kernel2 = np.zeros(kernel_l2.shape)
    
    # recording the loss in each epoch for plotting
    epoch_loss = []
    epoch_loss_test = []

    # velocities for momentum
    v_alpha = np.zeros(alpha.shape)
    v_beta = np.zeros(beta.shape)
    v_kernel1 = np.zeros(kernel_l1.shape)
    v_kernel2 = np.zeros(kernel_l2.shape)
    
    # cumulative gradients for rmsprop
    g_alpha = np.zeros(alpha.shape)
    g_beta = np.zeros(beta.shape)
    g_kernel1 = np.zeros(kernel_l1.shape)
    g_kernel2 = np.zeros(kernel_l2.shape)

    for e in range(epochs):
        if e == epochs-1:
            plt.figure(figsize=(10, 10))
            count_plot = 0
        np.random.shuffle(input_train)
        loss_train = 0
        loss_test = 0
        print('Epoch: %d' %(e+1), '/ %d' %epochs)
        for i in range(input_train.shape[0]):
            final, hidden, inp_mlp, inp_pool_2, inp_conv_2, inp_pool_1, inp_conv_1 = feed_forward(input_train[i][0], kernel_l1, kernel_l2, alpha, beta)

            y_act = np.array([1 if idx == input_train[i][1] else 0 for idx in range(10)])
            y_act = y_act.reshape((y_act.shape[0], 1))
            
            if np.all(final) == 0:
                loss_train += 0
            else:
                loss_train -= y_act.T @ np.log2(final)

            alpha_upd, beta_upd, gradient = mlp_backprop(inp_mlp, alpha, beta, hidden, final, y_act)
            
            update_alpha += alpha_upd
            update_beta += beta_upd

            gradient = pooling_backprop(inp_pool_2, gradient)
            

            upd_kerl2, gradient = conv_backprop(inp_conv_2, gradient, kernel_l2)
            update_kernel2 += upd_kerl2
            
            gradient = pooling_backprop(inp_pool_1, gradient)

            upd_kerl1, gradient = conv_backprop(inp_conv_1, gradient, kernel_l1)
            if e == epochs-1:
                count_plot += 1
                plt.subplot(10, 10, count_plot)
                plt.imshow((gradient - np.min(gradient))/(np.max(gradient - np.min(gradient))), cmap='hot')
                plt.xticks([])
                plt.yticks([])
                
            update_kernel1 += upd_kerl1
            
            if (i+1) % batch_size == 0:
                if variant == 'Vanilla':
                    alpha += learning_rate*update_alpha/batch_size
                    beta += learning_rate*update_beta/batch_size
                    kernel_l1 += learning_rate*update_kernel1/batch_size
                    kernel_l2 += learning_rate*update_kernel2/batch_size

                if variant == 'Momentum':
                    alpha, beta, kernel_l1, kernel_l2, v_alpha, v_beta, v_kernel1, v_kernel2 = momentum(alpha, beta, kernel_l1, kernel_l2, update_alpha, update_beta, update_kernel1, update_kernel2, v_alpha, v_beta, v_kernel1, v_kernel2, 0.9, learning_rate)
        
                if variant == 'RMSProp':
                    alpha, beta, kernel_l1, kernel_l2, g_alpha, g_beta, g_kernel1, g_kernel2 = rmsprop(alpha, beta, kernel_l1, kernel_l2, update_alpha, update_beta, update_kernel1, update_kernel2, g_alpha, g_beta, g_kernel1, g_kernel2, learning_rate)
        
                update_alpha = np.zeros(alpha.shape)
                update_beta = np.zeros(beta.shape)
                update_kernel1 = np.zeros(kernel_l1.shape)
                update_kernel2 = np.zeros(kernel_l2.shape)
        
        epoch_loss.append(loss_train[0][0]/input_train.shape[0])
        accu = 0
        # testing loss after each epoch
        for i in range(input_test.shape[0]):
            final_test, _, _, _, _, _, _ = feed_forward(input_test[i][0], kernel_l1, kernel_l2, alpha, beta)
            y_act_test = np.array([1 if idx == input_test[i][1] else 0 for idx in range(10)])
            y_act_test = y_act_test.reshape((y_act.shape[0], 1))
            
            if np.all(final) == 0:
                loss_test += 0
            else:
                loss_test -= y_act_test.T @ np.log2(final_test)
            if np.argmax(final_test) == input_test[i][1]:
                accu += 1
        print('Test Accuracy: %d' %accu, '%')
        epoch_loss_test.append(loss_test[0][0]/input_test.shape[0])
    plt.show()
    n_epochs = [(x+1) for x in range(epochs)]
    plt.title('CE Loss vs Epoch for variant: ' + str(variant))
    plt.plot(n_epochs, epoch_loss, label='Train Loss')
    plt.plot(n_epochs, epoch_loss_test, label='Test Loss')
    plt.grid(True)
    plt.legend()
    plt.show()
```
